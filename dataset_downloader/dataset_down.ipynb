```python
train_zip_path='train2017.zip'
vaild_zip_path='test2017.zip'
```

```python
os.system('unzip '+train_zip_path+' -d /train2017/')
os.system('unzip '+vaild_zip_path+' -d /vaild2017/')
os.system('mkdir raw')
```

```python

# train set
for train_zip_path in glob.glob('./train2017/*'):
	name=os.path.basename(train_zip_path)
	if 'gt' in name.split('_'):
		os.system('unzip '+train_zip_path+' -d ./raw/ch8_training_gt')
	else:
		os.system('unzip '+train_zip_path+' -d ./raw/ch8_training_images')
```

```python
# valid set

for vaild_zip_path in glob.glob('./valid2017/*'):
	name=os.path.basename(vaild_zip_path)
	if 'gt' in name.split('_'):
		os.system('unzip '+vaild_zip_path+' -d ./raw/ch8_validation_gt')
	else:
		os.system('unzip '+vaild_zip_path+' -d ./raw/ch8_validation_images')
```
